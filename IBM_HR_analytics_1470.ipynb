## Label Encoding  

In machine learning, we usually deal with datasets that contain multiple labels in one or more columns. These labels can be in the form of words or numbers.  
To make data understandable or in human-readable form, the training data is often labeled in words.  


```python
# Target Variable (Attrition)
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Encode binary variables
data['OverTime'] = data['OverTime'].map({'No': 0, 'Yes': 1})
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
```

```python
data['Over18'] = data['Over18'].map({'Y': 1, 'No': 0})
```

```python
# Encode categorical columns that are ordinal using LabelEncoder
from sklearn.preprocessing import LabelEncoder

encoding_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']

label_encoders = {}

for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

## Machine Learning: Splitting the data into Training and Testing sample

```python
X = data.drop('Attrition', axis=1)  # Input features
y = data['Attrition'].values  # Output
```

### Resampling

Resampling is the method that consists of drawing repeated samples from the original data samples. It is a nonparametric method of statistical inference.  

Oversampling and undersampling in data analysis are techniques used to adjust the class distribution of a dataset. These terms are used in statistical sampling, survey methodology, and machine learning.  

- **Oversampling** involves increasing the instances of the minority class to balance the dataset.
- **Undersampling** reduces the instances of the majority class to achieve balance.  

Both techniques help in handling class imbalances and improving model performance.


```python
# Should we oversample the minority class or undersample the majority class?

# There are two possible outputs for Attrition:
# - Attrition (10% of the data)
# - No Attrition (90% of the data)

# To balance this data, we have two approaches:

# - Oversampling â†’ Increase the minority class (Attrition) when we have a smaller dataset
#   and do not want to lose existing information.

# - Undersampling â†’ Reduce the majority class (No Attrition) when we have a lot of data
#   and can afford to lose some samples.

# Choosing the right approach depends on the dataset size and the importance of retaining data for model training.
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler

print("Before oversampling:", Counter(y))

rus = RandomOverSampler(random_state=42)
X_over, y_over = rus.fit_resample(X, y)

print("After oversampling:", Counter(y_over))
```

```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, test_size=0.2, random_state=42)
```

```python
# Sanity check for the sampled data
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Logistic Regression in Machine Learning  

Logistic Regression is used for predicting a category, especially binary categories (Yes/No, 0/1).  

For example:  
- Whether to approve a loan or not (Yes/No)?  
- Which group does a customer belong to (Silver/Gold/Platinum)?  

### Types of Logistic Regression:  
- **Binomial Logistic Regression:** When there are only two possible outcomes in the target variable.  
- **Multinomial Logistic Regression:** When there are more than two possible outcomes in the target variable.  
- **Ordinal Logistic Regression:** When the target variable has ordered categories (e.g., Small < Medium < Large).  


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
prediction = logreg.predict(X_test)
cnf_matrix = confusion_matrix(y_test, prediction)
print("Accuracy Score:", accuracy_score(y_test, prediction))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score

# Create figure
fig = plt.figure(figsize=(15, 6))

# Plot confusion matrix heatmap
ax1 = fig.add_subplot(1, 2, 1)
cnf_matrix = confusion_matrix(y_test, prediction)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap='Blues', fmt='d', ax=ax1)
plt.xlabel('Predicted')
plt.ylabel('Expected')

# Plot ROC curve
ax2 = fig.add_subplot(1, 2, 2)
y_pred_proba = logreg.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
auc = roc_auc_score(y_test, prediction)
plt.plot(fpr, tpr, label="AUC = " + str(auc))
plt.legend(loc=4)
plt.show()
```
